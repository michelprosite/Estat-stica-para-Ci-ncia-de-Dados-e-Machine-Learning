## Amostragem Aleatória Simples

```python
percente_amostra = int(len(dataset) * 0.2)
```

```python
def amostragem_aleatoria_simples(dataset, amostras):
    return dataset.sample(n = amostras, replace=False, random_state=1)
```

```python
df_amostragem_aleatoria_simples = amostragem_aleatoria_simples(dataset, percente_amostra)
df_amostragem_aleatoria_simples
```

## Amostra Sistemática

```python
def amostragem_sistematica(dataset, amostras):
    salt = len(dataset) // amostras
    random.seed(1)
    start = random.randint(0, salt)
    indices = np.arange(start, len(dataset), step=salt)
    amostragem_sistematica = dataset.iloc[indices]
    return amostragem_sistematica
```

```python
df_amostragem_sistematica = amostragem_sistematica(dataset, amostras=100)
df_amostragem_sistematica
```

## Amostragem por Gurpos

```python
def amostragem_agrupamento(dataset, numero_grupos):
    intervalo = len(dataset) / numero_grupos
    grupos = []
    id_grupo = 0
    contagem = 0
    for _ in dataset.iterrows():
        grupos.append(id_grupo)
        contagem += 1
        if contagem > intervalo:
            contagem = 0
            id_grupo += 1
    
    dataset['grupo'] = grupos
    random.seed(1)
    grupo_selecionado = random.randint(0, numero_grupos)
    return dataset[dataset['grupo'] == grupo_selecionado]
```

```python
df_amostragem_agrupamento = amostragem_agrupamento(dataset, 100)
df_amostragem_agrupamento
```
